```python
# Basic Libraries
import numpy as np  
import pandas as pd  

# Data Visualization
import matplotlib.pyplot as plt  
import seaborn as sns  

# Data Preprocessing
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import LabelEncoder, StandardScaler  

# Handling Missing Data
from sklearn.impute import SimpleImputer  

# Machine Learning (for later if needed)
from sklearn.ensemble import RandomForestRegressor  
from sklearn.metrics import mean_absolute_error, r2_score  
```

```python
# Display settings
pd.set_option('display.max_columns', None)  # Show all columns in output
sns.set_style("whitegrid")  # Set style for seaborn plots

print("All libraries imported successfully! ✅")
```

```python
df= pd.read_csv('Electric_Vehicle_Population_Data.csv') 
df.head()
```

```python
# Check for missing values in each column
df.isnull().sum()

```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Check if duplicates are removed
print(f"Number of rows after removing duplicates: {df.shape[0]}")
```

```python
df.describe()
```

```python
# Plot distribution of Electric Range
plt.figure(figsize=(10, 6))
sns.histplot(df['Electric Range'], bins=30, kde=True, color='blue')
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.show()

# Plot distribution of Base MSRP
plt.figure(figsize=(10, 6))
sns.histplot(df['Base MSRP'], bins=30, kde=True, color='green')
plt.title('Distribution of Base MSRP')
plt.xlabel('Base MSRP (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count of Electric Vehicle Types (BEVs vs PHEVs)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Electric Vehicle Type')
plt.title('Count of Electric Vehicle Types (BEVs vs PHEVs)')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.show()

# Count of EV Makes
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Make', order=df['Make'].value_counts().index[:10])  # Top 10 makes
plt.title('Top 10 EV Makes')
plt.xlabel('Count')
plt.ylabel('Make')
plt.show()
```

```python
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Compute the correlation matrix for the numerical columns only
correlation_matrix = df[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Count the number of vehicles in each county
county_counts = df['County'].value_counts()

# Show the top 10 counties with the most vehicles
print(county_counts.head(10))

# Plot the top 10 counties with the most vehicles
plt.figure(figsize=(12, 6))
county_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Counties with the Most Electric Vehicles')
plt.xlabel('County')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Proportion of vehicles by county (Pie chart)
county_proportion = df['County'].value_counts()

# Plot the pie chart for the top 10 counties
plt.figure(figsize=(8, 8))
county_proportion.head(10).plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3', 10), legend=False)
plt.title('Top 10 Counties with Proportion of Electric Vehicles')
plt.ylabel('')
plt.show()
```

```python
# Check the length of Electric Utility descriptions, handle NaN values and ensure string conversion
df['Electric Utility Length'] = df['Electric Utility'].fillna('').astype(str).apply(len)

# Show the unique lengths of Electric Utility descriptions
print(f"Unique lengths of Electric Utility descriptions: {df['Electric Utility Length'].unique()}")
```

```python
# Check for electric utilities that appear only once (possibly rare or typos)
rare_utilities = df['Electric Utility'].value_counts().tail(10)
print("Rare electric utilities (appearing only once):")
print(rare_utilities)
```

```python
# Example: Relationship between Electric Utility and Vehicle Location
utility_vs_location = df.groupby('Electric Utility')['Vehicle Location'].value_counts().unstack().fillna(0)
print("Electric Utility distribution across Vehicle Locations:")
print(utility_vs_location.head(10))  # Show top 10 electric utilities by vehicle location
```
